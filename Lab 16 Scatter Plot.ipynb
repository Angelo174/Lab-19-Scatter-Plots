
### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
missing_data = df['JobSatPoints_6'].isnull().sum()

print(f"Number of missing values in 'JobSatPoints_6': {missing_data}")
```

```python

# Calculate the mean of the 'JobSatPoints_6' column
mean_value = df['JobSatPoints_6'].mean()

# Replace missing values with the mean
df['JobSatPoints_6'].fillna(mean_value, inplace=True)

# Calculate the new mean after replacing missing values
new_mean_value = df['JobSatPoints_6'].mean()

# Round the new mean to one decimal place
rounded_mean = round(new_mean_value, 1)

print(f"The mean of 'JobSatPoints_6' rounded to one decimal place is: {rounded_mean}")
```

```python
## Write your code here
age = [25, 30, 35, 40, 45, 50, 55, 60]
job_satisfaction = [3, 4, 5, 4, 3, 4, 5, 4]

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(age, job_satisfaction, color='blue', alpha=0.5)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Job Satisfaction (Points)')
plt.title('Scatter Plot of Age vs. Job Satisfaction')

# Show the plot
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
!pip install seaborn
import seaborn as sns
```

```python
## Write your code here
# Replace these lists with your actual data
age = [25, 30, 35, 40, 45, 50, 55, 60]
job_satisfaction = [3, 4, 5, 4, 3, 4, 5, 4]

# Create a DataFrame if your data is not already in one
import pandas as pd
df = pd.DataFrame({'Age': age, 'JobSatPoints_6': job_satisfaction})

# Create a scatter plot with a regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='JobSatPoints_6', data=df, color='blue', scatter_kws={'alpha':0.5})

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Job Satisfaction (Points)')
plt.title('Scatter Plot with Regression Line for Age vs. Job Satisfaction')

# Show the plot
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here
compensation = [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000]
job_satisfaction = [3, 4, 5, 4, 3, 4, 5, 4]
age = [25, 30, 35, 40, 45, 50, 55, 60]

# Create a bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(compensation, job_satisfaction, s=age, color='purple', alpha=0.5)

# Add labels and title
plt.xlabel('Yearly Compensation')
plt.ylabel('Job Satisfaction (Points)')
plt.title('Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size')

# Show the plot
plt.show()
```
